## Summary
This study addresses the optimization problem of ***offshore wind farm site selection*** using a multi-objective evolutionary algorithm (***NSGA-III***), focusing specifically on the ***Kaliningrad region's Exclusive Economic Zone (EEZ)***. The optimization simultaneously considered three key criteria: maximizing ***average annual wind speed***, minimizing ***water depth***, and minimizing ***distance to the nearest port***, chosen due to their ***technical importance*** and ***availability of reliable public data***. ## **Focus of This Study: The Site Selection Problem**  

The ***selection of an optimal site*** is a critical challenge in offshore wind farm development. This process requires a careful balance between multiple factors, including:  

- ***Energy yield potential:*** Optimal wind conditions directly influence the efficiency and economic viability of the project, demanding detailed wind resource assessments and simulations based on meteorological data.

- ***Technical feasibility:*** Geological and geotechnical conditions influence foundation type and installation methods, significantly affecting project costs and risks.

- ***Economic considerations:*** The total cost, including construction, maintenance, grid connection, and energy transportation, must be minimized while maximizing expected revenue.

- ***Macro-Scale:*** *Determining the best general location for the wind farm within a large region.*  
- ***Micro-Scale:*** *Optimizing the precise arrangement of turbines within the selected site to enhance wind capture efficiency.*

## **Criteria and constraints for Offshore Wind Farm Site Selection within the EEZ**

***The Exclusive Economic Zone (EEZ)*** is a maritime area extending from the baseline from which the breadth of a nation's territorial sea is measured, outward to 200 nautical miles (approximately 370 kilometers) into the open sea. Within this zone, a state has specific sovereign rights and exclusive jurisdiction. These rights include the exploration, conservation, and management of natural marine resources, both living and non-living, located within the water column and on or beneath the seabed. Additionally, the state possesses the exclusive authority to construct, authorize, and regulate artificial islands and structures intended for economic exploitation, scientific research, and resource exploration, especially renewable energy production from wind, ocean currents, and wave energy.

When considering suitable areas for offshore wind farm (OWF) development within an EEZ, developers must comply with multiple environmental, technical, and socio-economic constraints. These constraints ensure the project's environmental sustainability, safety, economic viability, and compatibility with other maritime activities. Important criteria and constraints influencing site selection include (not in order of importance):

***Criteria***
- Wind resources;
- Depth, seabed geological characteristic;
- Proximity to the nearest port with sufficient infrastructure and capacity;
- Minimum distance to the existing high-voltage electrical grid infrastructure;
- Potential environmental impact;

***Constraints***
- Existing offshore wind farms (OWFs) and wave energy parks;
- Military operation or exercise zones;
- Presence of shipwrecks, unexploded ordnance (UXO), and other seabed obstructions;
- Vessel traffic routes, navigation lanes and harbor entrances;
- Areas for oil and gas exploration or production;
- Zones for extraction, dredging, and dumping activities;
- Areas important for commercial fishing;
- Pipelines (oil & gas) and cables (power & telecom) rights of way;
- Environmentally protected areas.

### **Multi-Objective Optimization in Offshore Wind Farm Site Selection**

The ***optimization of offshore wind farm site selection*** is a **multi-criteria** problem, meaning that it cannot be reduced to a single dominant criteria such as ***average wind speed***. While wind resource availability is one of the most critical factors for energy production, selecting an optimal location based solely on this parameter would be insufficient and potentially misleading. Other previously mentioned factors also play significant roles in determining the overall viability of a site. As a rule, these factors tend to be **conflicting with each other**. Therefore, compromises are necessary, requiring a ***balanced and sustainable approach.*** This complexity leads to the classification of the site selection process as a **multi-objective optimization problem**. 

**Multi-objective optimization (MOO)**, also known as **multi-criteria decision-making (MCDM)**, is a mathematical framework used to find solutions that optimize ***multiple conflicting objectives simultaneously***. Formally, a **multi-objective optimization problem** can be defined as follows:

*For convenience in notation and explanation, all objective functions are conventionally **formulated as minimization problems**. Even if a certain function is naturally more desirable at **higher values** (i.e., should be maximized), we simply **multiply it by (-1)** to convert it into a minimization problem. For example, we consider **average wind speed**, where a **higher value** indicates a **more favorable** location. However, to maintain consistency within the optimization framework, we will **invert the values (multiply by -1)** and work with negative values instead. This ensures a uniform approach across all objectives and simplifies the mathematical structure of the optimization process. The reasoning behind this transformation will become clearer when we examine the optimization model in detail.*

Flowchart below illustrates the process of **multi-objective optimization**, starting from *problem formulation and ending with the selection of an optimal solution*.  

1. **Formulating the Multi-Objective Optimization Problem**  
   - Multiple objective functions $f_1(x), f_2(x), ..., f_M(x)$ are optimized, typically aiming for **minimization** (maximization problems can also be converted into minimization for consistency).  
   - The solution is subject to ***constraints***, which can be represented as inequalities or equalities.  

2. **Applying Multi-Objective Optimization Techniques**  
   - Algorithms such as ***NSGA, MOEA/D, or Pareto-based methods*** are used to find an ***optimal set of solutions.***  
   - Unlike single-objective optimization, there is no single global optimum; instead, a ***set of Pareto-optimal solutions*** is identified.  

3. **Generating Multiple Optimal Solutions**  
   - The red curve with green points in the top-right block represents the ***Pareto front — a set of solutions where improving one objective results in trade-offs in another.***  

4. **Selecting a Single Optimal Solution**  
   - To finalize the decision, ***higher-level information*** is incorporated, such as expert preferences, decision-making criteria, or aggregation methods.  
   - This step helps select ***one preferred solution*** from the Pareto-optimal set.  

In this study, we use the ***Multi-Objective Optimization Technique NSGA-III*** to address the problem of optimizing multiple conflicting objectives.  

**NSGA-III (Non-dominated Sorting Genetic Algorithm III)** is an evolutionary multi-objective optimization algorithm (MOO) designed to solve problems with a large number of objective functions (typically more than three).

### **Stages of the NSGA-3 Algorithm**  
1) **Generation of the initial population** – random distribution of solutions in the decision space.  
2) **Evaluation of objective functions** – calculation of objective function values for all individuals.  
3) **Non-dominated sorting** – partitioning the population into dominance fronts.  
4) **Generation of reference directions** – defining a set of points uniformly distributed in a multi-dimensional space.  
5) **Selection of solutions for the next generation** – projecting solutions onto the reference directions.  
6) **Mutation and crossover operators** – standard genetic operators ensuring the evolution of solutions.  
7) **Repetition of steps 2-6** until the stopping criterion is met.

Now, I will provide ***a brief explanation*** of each stage of the algorithm.

---

#### 1) **Generation of the initial population**

The ***first step*** in the NSGA-III algorithm is the **generation of the initial population**, which consists of a set of potential solutions randomly distributed within the ***decision space***. This population serves as the ***starting point*** for the evolutionary optimization process and plays a crucial role in maintaining ***diversity*** and ensuring broad coverage of the search space.  

Each individual in the population represents a ***candidate solution*** to the optimization problem and is typically encoded as a vector of decision variables. These decision variables can be ***continuous, discrete, or mixed***, depending on the nature of the problem. 

**In our case**, we consider only ***two variables***, which are the ***coordinates of a point*** $(\text{latitude}, \text{longitude})$. This means that the first individual in the population will have the variables $\mathbf{ind}_1 = (\text{latitude}_1, \text{longitude}_1)$, the second individual will have $\mathbf{ind}_2 = (\text{latitude}_2, \text{longitude}_2)$, and so on. These ***two variables*** are sufficient to uniquely determine our objective functions, as for a given coordinate $\mathbf{ind}_i = (\text{latitude}_i, \text{longitude}_i)$, we can explicitly define the ***average wind speed, water depth, and distance to the nearest port.*** Since we are considering only these criteria, no additional variables are required.

The ***number of individuals*** in the initial population is usually predefined and denoted as ***N***, ensuring a sufficient level of diversity to prevent premature convergence.  

The initialization process is typically ***randomized*** to avoid bias and ensure that the search process explores various regions of the solution space.

After generating the initial population, each individual undergoes an **evaluation process** where its objective function values are calculated based on the optimization problem’s criteria.

---

#### 2) **Evaluation of objective functions**

After generating the initial population, the next step in the NSGA-3 algorithm is the **evaluation of objective functions**. This step is crucial as it determines the ***quality of each solution*** by calculating its ***objective values*** based on the given optimization problem. In our case, each solution is represented by a coordinate pair and the objective functions are evaluated for each individual in the population.  

Since we are considering ***three objective functions***, their evaluation is performed as follows:  

1. **Average Wind Speed ($f_1$)** – The wind speed at a given coordinate is retrieved from wind resource datasets, such as **Global Wind Atlas**.  

2. **Water Depth ($f_2$)** – The water depth at a given coordinate is retrieved from **GEBCO bathymetric data**.  

3. **Distance to the Nearest Port ($f_3$)** – This objective function is calculated as the ***geodesic distance*** from the candidate site to the nearest available port.  

For each candidate solution we compute the vector of objective values:

$$
F(ind_i) = (f_1(ind_i), f_2(ind_i), f_3(ind_i))
$$

where:  
- $f_1(ind_i) = \text{wind speed at } (\text{latitude}_i, \text{longitude}_i)$  
- $f_2(ind_i) = \text{absolute water depth at } (\text{latitude}_i, \text{longitude}_i)$  
- $f_3(ind_i) = \text{distance to the nearest port from } (\text{latitude}_i, \text{longitude}_i)$  

This evaluation provides the necessary information for **ranking solutions** in the next step, where they are sorted based on their dominance relations. Each individual is now associated with an objective vector, which will guide the selection process towards **Pareto-optimal solutions**.

---

#### 3) **Non-dominated Sorting** 

The **Non-dominated Sorting** step is a fundamental component of the NSGA-3 algorithm and plays an important role in ***ranking solutions*** based on ***Pareto dominance***. Since multi-objective optimization problems involve conflicting objectives, there is usually no single "best" solution. Instead, solutions are compared based on ***whether one dominates another***. The goal of this step is to ***classify individuals into different dominance levels (fronts)***, which ***helps in the selection process for the next generation.***  

##### **Definition of Pareto Dominance**  
A solution $ind_i$ is said to **dominate** another solution $ind_j$ if ***both*** conditions are true:
1. $ind_i$ is **no worse** than $ind_j$ in ***all objective functions***:  
   $$
   f_k(ind_i) \leq f_k(ind_j) \quad \forall k \in \{1,2, \dots, M\}
   $$

   For example, in our case, $ind_1$ has *greater or equal wind speed*, *lesser or equal water depth*, and *lesser or equal distance to the nearest port* compared to $ind_2$. **No worse** means better or equal
2. $ind_i$ is **strictly better** than $ind_j$ in ***at least one objective function***:  
   $$
   \exists k \text{ such that } f_k(x_i) < f_k(x_j)
   $$

A solution that is **not dominated by any other solution in the population** belongs to the **first Pareto front** also called **non-dominated Pateto-optimal set**. 

The aim of algorithm is to define this set of solutions.

##### **Steps of the Non-dominated Sorting Process**  
1. **Identifying the First Pareto Front (Best Solutions)**  
   - Identify solutions that are ***not dominated*** by any other individuals in the population.  
   - These solutions form the ***first Pareto front*** and are assigned ***rank = 1***.  

2. **Identifying the Second Pareto Front**  
   - Remove the solutions from the first front and find solutions that are ***not dominated*** within the remaining population.  
   - These solutions form the ***second Pareto front*** and are assigned ***rank = 2***.  

3. **Repeat the process**  
   - Continue this process until all solutions in the population have been assigned to a ***dominance front***.  
   - If a solution belongs to ***front $k$***, it means it is dominated only by solutions from ***fronts $( 1, 2, \dots, k-1)$***.  

Each front represents a different ***level of optimality***, with the first front containing the ***best trade-offs*** among the objective functions.

---

#### 4) **Generation of Reference Directions** 

The generation of reference directions is a key step in NSGA-3 that differentiates it from NSGA-2. Unlike NSGA-2, which relies on the crowding distance to maintain diversity among solutions, NSGA-3 introduces **reference directions**, which guide the selection of solutions and ensure **uniform coverage** of the Pareto front, especially in high-dimensional problems (when the number of objective functions is large, typically $M \geq 3$.  

##### **Why Are Reference Directions Needed?**  
When optimizing multiple conflicting objectives, the Pareto front can be complex and high-dimensional. Traditional dominance-based methods struggle to evenly distribute solutions across this front, ***leading to clustering in some areas while leaving others sparsely populated***. ***Reference directions help address this issue*** by:  
- ***Providing a structured way to distribute solutions*** across the Pareto front.  
- ***Ensuring diversity*** among selected solutions.  

##### **How Reference Directions Are Generated**  
The reference directions are a set of ***predefined points*** that are uniformly distributed within the objective space. These directions serve as ***anchors*** for projecting candidate solutions and are typically generated as follows:  

**Define the Number of Reference Directions**  
   - The number of reference directions depends on the number of objectives $M$ and the required resolution $H$.  
   - The total number of reference directions $N_r$ is calculated as:  
     $$
     N_r = \binom{H + M - 1}{M - 1}
     $$
   - Here, $H$ is a user-defined parameter that controls the granularity of the reference directions. A higher $H$ results in a finer distribution.  

Since we are working in ***a three-objective space***, the reference directions are generated in ***a 3D space***, ensuring a uniform distribution of potential solutions across the trade-off surface. ***This prevents the algorithm from favoring only one of the objectives and ensures a well-balanced selection of wind farm locations that consider all three criteria equally.***

#### 5) **Selection of Solutions for the Next Generation**  

After **3)non-dominated sorting** and **4)generation of reference directions**, NSGA-3 ***selects individuals for the next generation***. This selection process ensures that the ***best solutions*** are retained while maintaining a ***diverse distribution*** across the Pareto front. Unlike NSGA-2, which uses crowding distance for selection, NSGA-3 uses a **reference-based approach**, ensuring a more uniform spread of solutions in high-dimensional spaces.

##### **Key Steps in the Selection Process**  

1. **Selecting the Best Non-Dominated Solutions**  
   - Solutions are sorted into ***non-dominated fronts*** based on ***Pareto dominance***.  
   - The best solutions (those in the ***first Pareto front***) are automatically selected for the next generation.  
   - If the number of solutions in the first front ***exceeds the population size***, a ***reference-based selection mechanism*** is applied.  

2. **Projecting Solutions onto Reference Directions**  
   - Each remaining solution is projected onto the ***nearest reference direction*** in the objective space.  
   - The reference directions help ensure that the solutions are ***evenly distributed*** along the Pareto front.  

3. **Reference-Based Niching Mechanism**  
   - If too many solutions are clustered near a single reference direction, ***only a subset*** of them is selected to maintain diversity.  
   - Solutions that are ***closest to sparsely populated reference directions*** are preferred.  
   - This mechanism prevents the ***overcrowding*** of solutions in some areas while keeping underrepresented regions populated.  

4. **Choosing the Final Set of Solutions**  
   - Solutions are ***incrementally added*** to the next generation until the population size limit is reached.  
   - If multiple solutions are mapped to the same reference direction, the one ***closest to the reference point*** is chosen.  

By following this selection process, NSGA-3 ensures that the ***next generation of solutions*** maintains a diverse, high-quality representation of optimal trade-offs, leading to ***better convergence and more robust decision-making*** in multi-objective optimization problems.

#### 6) **Mutation and Crossover Operators in NSGA-3**  
 
In NSGA-3, as in other evolutionary algorithms, **mutation and crossover operators** play an important role in maintaining ***genetic diversity*** and ensuring the exploration of the solution space. These operators are applied to individuals selected for reproduction, generating new candidate solutions for the next generation.  

##### **Crossover Operator (Recombination of Parent Solutions)**  

The **crossover operator** combines genetic material from two parent solutions to create offspring. The goal of crossover is to ***explore new regions*** of the search space while inheriting beneficial traits from parent solutions. In NSGA-3, a common crossover method is the ***Simulated Binary Crossover (SBX)***, which is well-suited for real-valued decision variables.  

##### **Mutation Operator (Introducing Variability in Solutions)**  

Mutation introduces small random changes to an individual solution, ensuring that the algorithm does not get stuck in local optima. The most commonly used mutation method in NSGA-3 is ***Polynomial Mutation***.  

##### **Why Are Crossover and Mutation Important?**  

- ***Crossover enables exploration*** by recombining parent solutions to create new promising solutions.  
- ***Mutation ensures diversity*** and prevents premature convergence by introducing slight variations.  
- Together, these operators ***balance exploration and exploitation***, helping NSGA-3 efficiently search for high-quality solutions.  

In our ***offshore wind farm site selection problem***, these operators allow the optimization process to refine the locations of wind farms iteratively, ***adapting solutions*** to better balance wind speed, water depth, and distance to the nearest port.

#### 7) **Repetition of Steps 2-6 Until the Stopping Criterion is Met**  

The ***final step*** in NSGA-3 is the iterative repetition of the **main optimization cycle** (steps ***2-6***) until a predefined ***stopping criterion*** is met. Since NSGA-3 is an ***evolutionary algorithm***, it operates through ***multiple generations***, progressively refining the population of solutions towards the Pareto front. Each generation undergoes evaluation, selection, variation (mutation and crossover), and ranking until the algorithm converges to a high-quality set of solutions.  


##### **Stopping Criteria in NSGA-3**  
Since NSGA-3 is an iterative algorithm, it requires a condition to terminate execution. The stopping criterion is typically one of the following:  

1. **Maximum Number of Generations (Most Common Approach)**  
   - The algorithm runs for a fixed number of generations.
   - Once the number of iterations reaches this limit, the algorithm ***stops*** and returns the final population of Pareto-optimal solutions.  

2. **Convergence Criterion**  
   - The algorithm stops when ***improvements in solutions become negligible***.  
   - This is measured by tracking ***changes in the Pareto front*** over generations.  
   - If the movement of solutions between iterations falls below a threshold, the algorithm is considered to have converged.  

3. **Computational Budget Limit**  
   - The algorithm terminates when ***a pre-defined computational resource*** (e.g., CPU time, number of function evaluations) is exhausted.  
   - This approach is useful for real-time or large-scale problems.


##### **What Happens After the Algorithm Stops?**  
Once the stopping criterion is met, the ***final population of solutions*** represents the ***Pareto-optimal set = first Pareto front*** of this population found by NSGA-3. These solutions provide ***trade-offs between objectives***, allowing decision-makers to choose the most suitable compromise based on specific needs.  

In our **offshore wind farm site selection problem**, this means that after the final generation:  
- The algorithm outputs a ***set of optimal locations*** with different trade-offs between ***wind speed, water depth, and distance to the nearest port***.  
- ***Decision-makers*** can then analyze these options and ***select the best site based on additional real-world constraints or preferences***.

Now, we need to ***retrieve and store the EEZ data*** and convert it into a format that allows us to use the ***EEZ boundaries in Python***. This will enable us to perform spatial analysis, define constraints for optimization, and ensure that all selected offshore wind farm locations remain within the designated ***Exclusive Economic Zone (EEZ).***

***This code reads a GeoJSON file containing coordinates of EEZ bounds and creates a `Polygon` object from it. The resulting polygon can be used for spatial analysis in geographic applications.***

```python
"""Load a polygon (EEZ of Kaliningrad region) from a GeoJSON file and create a Polygon object using Shapely."""

import json
import numpy as np
from shapely.geometry import Polygon, Point

# Define the path to the GeoJSON file
FILE_PATH = r"C:\Users\Георгий\Desktop\MOOP Offshore Site Selection\Bounds_of_EEZ.geojson"

# Load the GeoJSON file
with open(FILE_PATH, "r", encoding="utf-8") as file:
    data = json.load(file)

# Extract polygon coordinates from the first feature
polygon_coords = data["features"][0]["geometry"]["coordinates"][0]

# Create a Polygon object using Shapely
polygon = Polygon(polygon_coords)

# Get the min/max values of the polygon boundaries
min_longitude, min_latitude, max_longitude, max_latitude = polygon.bounds
```

Next, the obtained ***EEZ polygon*** needs to be ***evenly filled with points***. This step is essential for further analys and the ***purpose of this step*** will become clearer as we progress through the practical implementation. 

Through several trials and iterations, a ***step size of 0.03*** was chosen to ensure ***sufficient accuracy***, which is approximately equivalent to ***1.9 km***. This means that our points will be ***evenly distributed*** within the ***EEZ boundaries*** with a spacing of ***1.9 km*** between them.

```python
"""Generate a grid of points within the polygon boundaries and filter only those inside the polygon."""
import numpy as np
# Set the grid step (smaller value = more points = more accuracy, but more complexity)
grid_step = 0.03  # approximately 1.9 km step

# Generate grid points
longitude_coords = np.arange(min_longitude, max_longitude, grid_step)
latitude_coords = np.arange(min_latitude, max_latitude, grid_step)
grid_points = np.array(np.meshgrid(longitude_coords, latitude_coords)).T.reshape(-1, 2)

points_inside_polygon = []

# Filter points that are inside the polygon
for point in grid_points:
    grid_point = Point(point)
    if polygon.contains(grid_point):
        points_inside_polygon.append(grid_point)

# Convert the list of POINT objects to a numpy array
points_inside_polygon_array = np.array([[point.x, point.y] for point in points_inside_polygon])
```

As a result, we obtained ***1,541 points*** that are ***evenly distributed*** within the ***EEZ boundaries.***

Our ***first objective function*** in the optimization process is the ***average wind speed***. Therefore, our task is to ensure that we can ***calculate the average wind speed at any point within the EEZ***. We need to develop a code that, given an ***input coordinate (latitude, longitude)***, returns the ***average wind speed value*** at that location.

The code below performs ***wind speed interpolation*** using ***the K-Nearest Neighbors (KNN) regression model***. Below is a breakdown of its functionality:

1. **Reading Training Data (`wind_data.txt`)**  
   - The code loads wind speed data from a file located at `"C:\Users\Георгий\Desktop\MOOP Offshore Site Selection\wind_data.txt"`.  
   - It extracts latitude, longitude, and wind speed values. 
   - These values are stored in `X_train` (coordinates) and `y_train` (wind speeds).

2. **Defining a Test Point**  
   - A single test coordinate `(..., ...)` is set in `X_test`, where wind speed needs to be predicted.

3. **Training the KNN Model**  
   - A ***K-Nearest Neighbors Regressor*** (`KNeighborsRegressor`) is initialized with `n_neighbors=2` and `weights="distance"`.
   - The parameter `n_neighbors=2` (`2` was chosen as optimal and effective after evaluating different values and their impact on prediction accuracy) ensures that the prediction is based on the ***two closest known points***, preventing extreme extrapolation while capturing local variations. The `weights="distance"` setting assigns ***higher influence to closer neighbors***, ensuring smoother interpolation of wind speed values within the EEZ.  
   - The model is trained using the dataset `X_train` (coordinates) and `y_train` (wind speeds).

4. **Predicting Wind Speed for the Test Point**  
   - The trained KNN model predicts the wind speed for the specified test coordinate.  
   - The result is rounded to two decimal places and stored in `predicted_wind_speeds`.

*This approach allows estimating wind speed at any given location based on the nearest known data points, weighted by distance.*

```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor

# === 1. Load wind data from wind_data.txt (training data) ===
wind_data_path = r"C:\Users\Георгий\Desktop\MOOP Offshore Site Selection\wind_data.txt"
train_coordinates = []
train_wind_speeds = []

# Read wind data and extract coordinates (latitude, longitude) and wind speed values
with open(wind_data_path, 'r') as file:
    for line in file:
        lat, lon, wind_speed = map(float, line.strip().split(','))
        train_coordinates.append([lat, lon])
        train_wind_speeds.append(wind_speed)

# Convert lists to NumPy arrays
X_train = np.array(train_coordinates)
y_train = np.array(train_wind_speeds)

# Define a test location for wind speed prediction
X_test = [(55.063274, 19.472847)]

# === 2. Train the KNN model ===
knn = KNeighborsRegressor(n_neighbors=2, weights="distance")  # Use 2 nearest neighbors with distance-based weighting
knn.fit(X_train, y_train)

# === 3. Predict wind speed for the test location ===
predicted_wind_speeds = np.round(knn.predict(X_test), 2)

# Print the predicted wind speed for the test location
print(f"Predicted wind speed at {X_test[0]}: {predicted_wind_speeds[0]} m/s")
```

#### **Overview of `scikit-learn` and K-Nearest Neighbors (KNN) Regression**

`scikit-learn` is one of the most widely used ***machine learning libraries*** in Python. It provides simple and efficient tools for ***data mining, analysis, and modeling***, covering various supervised and unsupervised learning algorithms.

One of the key models in `scikit-learn` is ***K-Nearest Neighbors (KNN)***, which can be used for both ***classification and regression*** tasks. KNN regression estimates the output for a given input by averaging the values of its nearest neighbors.

#### Example: KNN Regression in `scikit-learn`
The following example demonstrates ***K-Nearest Neighbors Regression*** using the `KNeighborsRegressor` class from `sklearn.neighbors`. The model is trained on a ***1D dataset***, and predictions are made for unseen values.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

# Generate sample data
np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)  # 40 random points in range [0,5]
y = np.sin(X).ravel()  # Target values using sine function
y[::5] += 0.5 * (np.random.rand(8) - 0.5)  # Add noise to every 5th point

# Define test data for predictions
X_test = np.linspace(0, 5, 500).reshape(-1, 1)

# Create and fit the KNN regressor models
n_neighbors = 5  # Define the number of neighbors
knn_uniform = KNeighborsRegressor(n_neighbors=n_neighbors, weights='uniform')
knn_distance = KNeighborsRegressor(n_neighbors=n_neighbors, weights='distance')

knn_uniform.fit(X, y)
knn_distance.fit(X, y)

# Predict using both models
y_pred_uniform = knn_uniform.predict(X_test)
y_pred_distance = knn_distance.predict(X_test)

# Plot the results
plt.figure(figsize=(10, 5))

plt.scatter(X, y, c='black', label="Training Data")  # Training points
plt.plot(X_test, y_pred_uniform, color='blue', label="KNN (Uniform Weights)")
plt.plot(X_test, y_pred_distance, color='red', linestyle='dashed', label="KNN (Distance Weights)")

plt.xlabel("X")
plt.ylabel("Target Value")
plt.title(f"KNN Regression (n_neighbors={n_neighbors})")
plt.legend()
plt.show()
```

Additionally, we will evaluate the results using the following performance metrics:

- **Mean Absolute Error (MAE):**  
  Measures the ***average absolute difference*** between predicted and actual wind speed values. Lower values indicate better accuracy.  
  $$
  \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} | y_i - \hat{y}_i |
  $$

- **Mean Squared Error (MSE):**  
  Calculates the ***average squared difference*** between predicted and actual values. Squaring the errors penalizes larger deviations more than smaller ones.  
  $$
  \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
  $$

- **Root Mean Squared Error (RMSE):**  
  The square root of MSE, which provides an error measure in the same unit as the wind speed (m/s). It is more sensitive to larger errors than MAE.  
  $$
  \text{RMSE} = \sqrt{\text{MSE}}
  $$

- **R-Squared $R^2$ Score:**  
  Measures how well the model explains the variance in the actual wind speed values. It ranges from ***0 to 1***, where values closer to ***1*** indicate a better fit.  
  $$
  R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
  $$

By analyzing these metrics and visualizing the comparison, we can assess how well our model predicts wind speeds at new locations.

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

mae_wind = mean_absolute_error(y_test, predicted_wind_speeds)

mse_wind = mean_squared_error(y_test, predicted_wind_speeds)
rmse_wind = np.sqrt(mse_wind)

r2_wind = r2_score(y_test, predicted_wind_speeds)

print(f"MAE: {mae_wind:.6f}")
print(f"RMSE: {rmse_wind:.4f} m/s")
print(f"R²: {r2_wind:.4f}")
```

### **Finding the Shortest Path from the Port to the Target Point**
  
This code aims to determine the ***shortest navigable path*** from a ***port*** to a ***target location*** by considering only predefined available points inside the EEZ. The path is constructed using ***graph-based navigation***, where edges represent ***distances between neighboring points***.  

The breakdown of the code is as follows:

---

### **1. Defining Key Locations**
- **Port Coordinates (`port_coords`)**  
  The port serves as the ***starting point*** for the pathfinding process.  
- **Target Coordinates (`target_point`)**  
  The target represents the ***final destination*** within the EEZ.


### **2. Graph Construction (`NetworkX`)**
- A **graph (`G`)** is initialized where:
  - Each ***point*** inside the EEZ is a ***node***.
  - Edges represent ***distances between neighboring points***, ensuring connectivity.


### **3. Constructing a KDTree for Fast Neighbor Search**
- A ***KDTree*** is built using all available points inside the EEZ.
- The tree allows ***fast nearest-neighbor lookup***, avoiding brute-force distance calculations.


### **4. Connecting Each Point to Its Nearest Neighbors**
- Each point is linked to its ***7 closest neighbors (`k_nearest=7`)*** to ensure connectivity.
- ***The edge weight*** (distance) is computed using the ***Haversine formula***, which calculates real-world distances between geographic coordinates.


### **5. Adding the Port as a Node**
- The ***port is added to the graph*** as a node labeled `"port"`.
- The ***port is connected to its 10 nearest points*** inside the polygon to ensure smooth navigation.


### **6. Finding the Nearest Available Point to the Target**
- Since the ***target location might not be inside `points_inside_polygon_array`***, we ***find the closest available point*** using the KDTree.


### **7. Finding the Shortest Path**
- ***Dijkstra’s algorithm*** is used to determine the ***shortest path*** between the ***port*** and the ***nearest available target point***.
- The ***shortest path distance*** is computed by summing up the distances along each step.


### **8. Computing the Total Travel Distance**
- The total distance is computed by summing up the ***stepwise distances*** between consecutive points in the shortest path.


### **Summary**
This approach ensures that ***only valid navigable points*** are considered while avoiding direct travel over restricted areas.  
It provides an ***efficient way*** to calculate the ***optimal route*** for offshore wind farm logistics, minimizing unnecessary travel distance while ensuring realistic navigation.

### **Ensuring All Points Remain Within the appropriate Polygon**  

In the optimization process for offshore wind farm site selection, it is crucial that all candidate locations remain ***inside the research area***. This ensures compliance with regulatory constraints and avoids infeasible solutions. However, when performing ***evolutionary operations***, some points may fall outside the valid area.  

The code below implements a ***repair mechanism*** using the `PolygonRepair` class. This class ensures that every solution remains ***within the predefined polygon boundary***, replacing any ***invalid points*** with randomly generated alternatives inside the polygon.


### **Functionality Breakdown**

1. **Defining the `PolygonRepair` Class**  
   - This class inherits from `pymoo`'s `Repair` class, making it compatible with evolutionary algorithms.
   - It takes a ***shapely polygon*** as input, which defines the valid boundary.

2. **Generating Random Points Inside the Polygon**  
   - The method `generate_random_point_inside()` is responsible for creating ***random valid points***.  
   - It repeatedly generates random `(latitude, longitude)` coordinates within the ***polygon's bounding box***, checking whether they lie inside the polygon.  
   - If the point is ***valid***, it is returned; otherwise, the process continues until a valid point is found.

3. **Repairing Invalid Points (`_do` Method)**  
   - The function iterates over the input points and checks whether they ***lie inside the polygon***.
   - If a point is ***valid***, it remains unchanged.
   - If a point is ***outside***, it is ***replaced with a randomly generated valid point*** from `generate_random_point_inside()`.
   - The repaired set of points is returned as a NumPy array.

*By implementing this Repair function, we can confidently optimize wind farm locations while ensuring that all candidate points remain inside the research space.*

```python
import numpy as np
from shapely.geometry import Point
from pymoo.core.repair import Repair

class PolygonRepair(Repair):
    def __init__(self, polygon):
        """
        Initialize the repair operator with the given polygon.
        
        Parameters:
        polygon (shapely.geometry.Polygon): The polygon defining valid boundaries
        """
        super().__init__()
        self.polygon = polygon

    def generate_random_point_inside(self):
        """Generate a random point inside the polygon."""
        min_x, min_y, max_x, max_y = self.polygon.bounds  # Get bounding box of the polygon
        
        while True:
            rand_x = np.round(np.random.uniform(min_x, max_x), 2)  # Generate random longitude
            rand_y = np.round(np.random.uniform(min_y, max_y), 2)  # Generate random latitude
            point = Point(rand_x, rand_y)
            
            if self.polygon.contains(point):  
                return [rand_x, rand_y]  # Return valid coordinates inside the polygon

    def _do(self, problem, X, **kwargs):
        """
        Repair method to ensure all points are inside the polygon.
        
        Parameters:
        X (numpy.ndarray): Array of points to be checked and repaired if necessary.
        
        Returns:
        numpy.ndarray: Repaired array of points, all within the polygon.
        """
        repaired_X = []
        
        for x in X:
            point = Point(x[1], x[0])  # Convert (lat, lon) to (lon, lat) format for consistency
            
            if self.polygon.contains(point):
                # If the point is inside the polygon, keep it unchanged
                repaired_X.append(x)
            else:
                # If the point is outside, replace it with a randomly generated valid point
                new_x, new_y = self.generate_random_point_inside()
                repaired_X.append([new_y, new_x])  # Convert back to (lat, lon) format
        
        return np.array(repaired_X)
```

### **Custom Sampling Strategy for Constrained Optimization**
In ***multi-objective optimization***, sampling within a constrained region is crucial for ensuring valid candidate solutions. This code defines a custom sampling method for generating random points inside a ***predefined polygon***.

### **Functionality Breakdown**

1. **Defining the `CustomSampling` Class**  
   - This class extends `pymoo`'s `Sampling` class, making it suitable for evolutionary optimization algorithms.  
   - It takes a polygon as input, which defines the spatial constraints for valid sampling.  
   - The ***purpose*** of this class is to ***generate sample points that strictly remain inside the polygon***.

2. **Extracting Polygon Boundaries**  
   - The function retrieves the minimum and maximum latitude/longitude values of the polygon.  
   - These boundaries form a ***bounding box***, which serves as the search area for generating sample points.  

3. **Generating Random Points Inside the Polygon**  
   - The function generates a large number of random points within the bounding box.  
   - This ensures that there are sufficient valid points to extract from.  

4. **Filtering and Validating Sampled Points**  
   - Each generated point is checked to determine whether it falls inside the polygon.  
   - If a point is ***valid***, it is ***stored*** for further use.  
   - If a point is ***outside*** the polygon, it is ***discarded***.  

5. **Ensuring the Required Number of Samples**  
   - The process continues until the ***required number of points is collected***.  
   - This guarantees that the function returns the exact number of requested samples.  

6. **Returning the Sampled Points in the Correct Format**  
   - The final list of valid points is converted into a structured array.  
   - The coordinate order is adjusted for consistency with geospatial data formats.  

*The class below ensures that every sampled point remains inside the polygon, preventing infeasible solutions that could occur if points were generated in unrestricted areas.*

```python
from pymoo.core.sampling import Sampling

class CustomSampling(Sampling):
    def __init__(self, polygon):
        """
        Initialize the custom sampling class with a given polygon.

        Parameters:
        polygon (shapely.geometry.Polygon): The polygon that defines the valid region for sampling.
        """
        super().__init__()
        self.polygon = polygon

    def _do(self, problem, n_samples, **kwargs):
        """
        Generate valid random samples inside the given polygon.

        Parameters:
        n_samples (int): The number of required sample points.

        Returns:
        np.ndarray: An array of sampled points inside the polygon, formatted as (latitude, longitude).
        """
        xl, yl, xu, yu = self.polygon.bounds  # Get the bounding box of the polygon
        X = []  # List to store valid sampled points

        while len(X) < n_samples:
            # Generate multiple random points at once
            samples = np.random.uniform([xl, yl], [xu, yu], size=(n_samples * 2, 2))

            for x in samples:
                if self.polygon.contains(Point(x[0], x[1])):  
                    X.append([np.round(x[0], 2), np.round(x[1], 2)]) # Add valid points inside the polygon
                if len(X) >= n_samples:
                    break  # Stop generating when the required number of points is reached

        return np.array(X)[:, [1, 0]]  # Return points as (latitude, longitude) format
```

### **Core Algorithm Framework**
The code below defines a ***multi-objective optimization problem*** for selecting the best offshore wind farm locations. It evaluates potential sites based on ***three key factors***: maximizing wind speed, minimizing water depth (shallower areas are preferred), and minimizing the distance to the nearest port for logistical efficiency.

### **Functionality Breakdown**

1. **Defining the `OffshoreWindFarmProblem` Class**  
   - This class inherits from `pymoo`'s `Problem` class, enabling multi-objective optimization for offshore wind farm site selection.  
   - The objective is to find locations that maximize wind speed, minimize water depth, and reduce distance to the port.  

2. **Building the Graph for Shortest Path Calculation**  
   - The EEZ region is represented as a graph where nodes are valid candidate points, and edges connect neighboring nodes.  
   - The `build_graph()` function constructs the graph using a `KDTree`, linking each point to its nearest neighbors.  

3. **Training KNN Models for Wind Speed & Depth Estimation**  
   - Two separate ***K-Nearest Neighbors (KNN) regression models*** are trained using real wind speed and depth datasets.  
   - These models predict wind speed and depth at any given location within the EEZ.  

4. **Computing Distance to the Port**  
   - The function `calculate_distance_to_port()` determines the shortest path from any candidate site to the port.  
   - If ***no valid path exists***, the function assigns an ***infinite distance***.  
   - This ensures that isolated or unreachable locations are automatically penalized in the optimization.

5. **Evaluating the Candidate Locations**  
   - The `_evaluate()` method computes three objectives:  
     ***Wind Speed*** (higher is better)  
     ***Water Depth*** (shallower is better)  
     ***Distance to the Port*** (shorter is better)  
   - The results are stored as a NumPy array and fed into the optimization algorithm.


*All previously mentioned functions are implemented within the `OffshoreWindFarmProblem` class. This structure allows integration with ***multi-objective evolutionary algorithms*** for optimizing offshore wind farm locations.*  


```python
import networkx as nx
from pymoo.core.problem import Problem
from sklearn.neighbors import KNeighborsRegressor
from haversine import haversine, Unit
from scipy.spatial import KDTree

class OffshoreWindFarmProblem(Problem):
    def __init__(self, wind_data_path, elevation_data_path, port_coords, points_inside_polygon_array):
        """
        Initialize the multi-objective optimization problem for offshore wind farm site selection.

        Parameters:
        wind_data_path (str): Path to the file containing wind speed data.
        elevation_data_path (str): Path to the file containing depth (elevation) data.
        port_coords (tuple): Latitude and longitude of the port.
        points_inside_polygon_array (numpy.ndarray): Array of valid points inside the EEZ polygon.
        """
        super().__init__(
            n_var=2,      # Number of decision variables (latitude and longitude of wind turbine locations)
            n_obj=3,      # Number of objective functions (wind speed, depth, and distance to port)
            n_constr=0,   # Number of constraints (not explicitly defined in this problem)
            xl=[54.4585, 18.9316],  # Lower bounds for latitude and longitude (min values within EEZ)
            xu=[55.9343, 20.9507]   # Upper bounds for latitude and longitude (max values within EEZ)
        )


        # Store port coordinates and EEZ nodes
        self.port_coords = port_coords
        self.points_inside_polygon_array = points_inside_polygon_array
        self.tree = KDTree(points_inside_polygon_array)

        # Create a graph for shortest path computation
        self.G = nx.Graph()
        self.build_graph()

        # Train KNN models for wind speed and depth estimation
        self.knn_wind = self.train_knn_model(wind_data_path)
        self.knn_elevation = self.train_knn_model(elevation_data_path)

    def build_graph(self):
        """
        Construct a graph where nodes represent points inside the EEZ, and edges connect nearest neighbors.

        This is used to compute the shortest distance to the port.
        """
        # Add nodes (all available points in EEZ)
        for idx, point in enumerate(self.points_inside_polygon_array):
            self.G.add_node(idx, pos=(point[1], point[0]))  # (longitude, latitude)

        # Define edges between each point and its nearest neighbors
        k_nearest = 7
        for i, point in enumerate(self.points_inside_polygon_array):
            distances, indices = self.tree.query(point, k=k_nearest + 1)
            for j in indices[1:]:  # Skip the first index (self-reference)
                self.G.add_edge(i, j, weight=haversine(
                    point, self.points_inside_polygon_array[j], unit=Unit.KILOMETERS))

        # Add the port as a node
        self.G.add_node("port", pos=self.port_coords)

        # Connect the port to its 10 nearest EEZ points
        port_distances, port_indices = self.tree.query([self.port_coords[::-1]], k=10)
        for idx in port_indices[0]:
            self.G.add_edge("port", idx, weight=haversine(
                self.port_coords, self.points_inside_polygon_array[idx], unit=Unit.KILOMETERS))

    def train_knn_model(self, data_path):
        """
        Train a K-Nearest Neighbors (KNN) regression model for wind speed or depth prediction.

        Parameters:
        data_path (str): Path to the file containing latitude, longitude, and the target variable.

        Returns:
        KNeighborsRegressor: Trained KNN regression model.
        """
        coordinates, targets = [], []
        with open(data_path, 'r') as file:
            for line in file:
                lat, lon, value = map(float, line.strip().split(','))
                coordinates.append([lat, lon])
                targets.append(value)

        X_train = np.array(coordinates)
        y_train = np.array(targets)

        knn = KNeighborsRegressor(n_neighbors=2, weights="distance")
        knn.fit(X_train, y_train)
        return knn

    def calculate_distance_to_port(self, points):
        """
        Compute the shortest distance from a given set of points to the port using the graph.

        Parameters:
        points (numpy.ndarray): Array of candidate site coordinates.

        Returns:
        numpy.ndarray: Array of shortest path distances to the port.
        """
        distances = []
        for target_point in points:
            # Find the nearest EEZ node to the target point
            _, nearest_target_idx = self.tree.query([target_point[::-1]], k=1)
            nearest_target_node = nearest_target_idx[0]

            # Compute the shortest path to the port
            try:
                path = nx.shortest_path(self.G, source="port", target=nearest_target_node, weight="weight")
                step_distances = [
                    haversine(self.G.nodes[path[i]]['pos'], self.G.nodes[path[i + 1]]['pos'], unit=Unit.KILOMETERS)
                    for i in range(len(path) - 1)
                ]
                total_distance = sum(step_distances)
            except nx.NetworkXNoPath:
                total_distance = float('inf')  # If no path exists, set distance to infinity

            distances.append(total_distance)
        return np.array(distances)

    def _evaluate(self, X, out, *args, **kwargs):
        """
        Evaluate the objective functions for the given set of candidate wind farm locations.

        Parameters:
        X (numpy.ndarray): Array of candidate site coordinates (latitude, longitude).
        out (dict): Output dictionary where objective function values will be stored.
        """
        # Predict wind speeds and depth using KNN models
        wind_speeds = np.round(self.knn_wind.predict(X), 2)
        elevations = np.round(self.knn_elevation.predict(X), 2)

        # Compute shortest distances to the port
        distances_to_port = np.round(self.calculate_distance_to_port(X),3)

        # Store the three objective function values
        out['F'] = np.column_stack([
            -wind_speeds,       # Maximize wind speed (minimizing negative wind speed)
            -elevations,        # Minimize depth (original values are negative, so we invert the sign to ensure shallower locations are preferred)
            distances_to_port   # Minimize distance to the port
        ])

# Files paths and port coordinates
wind_data_path = r"C:\Users\Георгий\Desktop\MOOP Offshore Site Selection\wind_data.txt"
elevation_data_path = r"C:\Users\Георгий\Desktop\MOOP Offshore Site Selection\elevation_data.txt"
port_coords = (54.65038, 19.86139)  # Coordinates of the offshore port

# Initialize the problem
problem = OffshoreWindFarmProblem(wind_data_path, elevation_data_path, port_coords, points_inside_polygon_array)
```

### **Optimization of Offshore Wind Farm Locations using NSGA-III**

The code below implements the ***NSGA-III (Non-dominated Sorting Genetic Algorithm III)*** to find the most suitable locations for offshore wind farms.

### **Functionality Breakdown**

1. **Setting Up Reference Directions for NSGA-III**  
   - Since we are optimizing ***three objectives***, we generate ***reference directions*** using the ***Das-Dennis*** method.  
   - These reference directions help NSGA-III maintain diversity in solutions by guiding the evolutionary process.  

2. **Initializing the NSGA-III Algorithm**  
   - The algorithm is configured with:  
     - **Custom Sampling (`CustomSampling`)**: Ensures that the initial population is generated within the predefined polygon.  
     - **Repair Function (`PolygonRepair`)**: Fixes any points that fall outside the predefined polygon.  
     - **Population Size**: `pop_size` individuals per generation (automatically equal to number of reference directions).  
     - **Mutation (`PM`)**: ***Polynomial mutation (PM)***.
     - **Crossover (`SBX`)**: ***Simulated Binary Crossover (SBX)***.

3. **Running the Optimization**  
   - The ***minimize()*** function executes the optimization process with the following parameters:  
     - **Generations (`n_gen`)** to allow sufficient evolution.  
     - **Random Seed (`seed=1`)** ensures reproducibility of results for further comparing.  
     - **History Storage (`save_history=True`)** tracks optimization progress over generations.  
     - **Verbose Output (`verbose=True`)** provides real-time logging of the optimization process.  

4. **Extracting the Results**  
   - The optimized ***decision variables (`X`)*** represent the best locations for wind turbines.  
   - The corresponding ***objective values (`F`)*** provide insight into wind speed, depth, and distance-to-port trade-offs.  

The optimization process was executed multiple times with different values for **`n_partitions`**, **`SBX`**, **`PM`**, **`pop_size`**, and **`n_gen`**.  

*Through experimentation, the ***most effective and well-representative configuration*** was selected, balancing ***solution diversity***, ***convergence speed***, and ***optimization accuracy***.*

```python
from pymoo.algorithms.moo.nsga3 import NSGA3
from pymoo.optimize import minimize
from pymoo.util.ref_dirs import get_reference_directions
from pymoo.operators.crossover.sbx import SBX
from pymoo.operators.mutation.pm import PM 

class RoundedSBX(SBX):
    def _do(self, problem, X, **kwargs):
        X = super()._do(problem, X, **kwargs)
        return np.round(X, 2)  

class RoundedPM(PM):
    def _do(self, problem, X, **kwargs):
        X = super()._do(problem, X, **kwargs)
        return np.round(X, 2)  

# Setting up reference directions for a 3-objective optimization problem. The parameter `n_partitions` determines the granularity of the reference directions
ref_dirs = get_reference_directions("das-dennis", 3, n_partitions=20)

# Initializing the NSGA-III algorithm
algorithm = NSGA3(ref_dirs=ref_dirs,
                  sampling=CustomSampling(polygon_without_territorial),  # Custom sampling method to ensure points are inside the polygon
                  repair=PolygonRepair(polygon_without_territorial),  # Repair function to fix out-of-bounds points
                  mutation=RoundedPM(eta=10), # Polynomial mutation
                  #pop_size=100,  
                  crossover=RoundedSBX())  # Simulated Binary Crossover

# Running the optimization process
res = minimize(problem,  # The offshore wind farm site selection problem
               algorithm,  # The configured NSGA-III algorithm
               termination=('n_gen', 100),  # The optimization runs for (n_gen=...) generations
               seed=1,  # Set a random seed for reproducibility
               save_history=True,  # Save the optimization history
               verbose=True)  # Enable detailed logging of the optimization process

# Extracting the final solutions and corresponding objective values
X = res.X  # Optimized decision variables (selected site coordinates)
F = res.F  # Corresponding objective function values (wind speed, depth, and distance to port)
```

### **Column Breakdown:**
1. **n_gen** – The current generation (iteration) of the algorithm.
2. **n_eval** – The total number of function evaluations performed at this iteration.
3. **n_nds** – The number of ***non-dominated solutions (NDS)*** in the ***Pareto front*** (how many solutions remain at the first level of optimality).
4. **eps** – The ***epsilon value*** (`ε`), representing the difference between the previous and current Pareto front. A smaller value indicates a more stable solution.
5. **indicator** – Shows the ***quality or status*** of the optimization process.

This output helps evaluate the ***convergence*** of the algorithm and determines how quickly the Pareto front stabilizes.

```python
# Compute the ideal point (minimum values for each objective)
approx_ideal = F.min(axis=0)

# Compute the nadir point (maximum values for each objective)
approx_nadir = F.max(axis=0)

# Normalize the objective values to the range [0,1] and round to 2 decimal places
nF = np.round((F - approx_ideal) / (approx_nadir - approx_ideal), 2)

print("Ideal Point (Minimum values for each objective):\n", approx_ideal)
print("\nNadir Point (Maximum values for each objective):\n", approx_nadir)
```

### **Hypervolume (HV) Metric for Pareto Front Evaluation**  

***Hypervolume (HV)*** is one of the most widely used metrics for assessing the ***quality of a Pareto front*** in multi-objective optimization. It measures the ***volume in the objective space*** that is ***dominated*** by the obtained solutions, given a predefined ***reference point***. A ***higher hypervolume value*** indicates a ***better and more well-distributed Pareto front***.  

#### **How is Hypervolume Calculated?**  
- A ***reference point*** is chosen, typically ***worse*** than all obtained solutions (e.g., the worst observed values in each objective function).  
- The ***Pareto front solutions*** define a region in the multi-objective space that is ***dominated*** by the front.  
- The ***hypervolume*** is the total space enclosed between the Pareto front and the reference point.  

By computing the ***hypervolume***, we can assess how well the ***60 solutions*** cover the ***trade-off space***, helping us determine the ***most optimal sites for offshore wind farm placement***.

```python
from pymoo.indicators.hv import HV

# Approximate ideal point: the best (minimum) value for each objective in the final generation
approx_ideal_all_population = res.history[len(res.history)-1].pop.get("F").min(axis=0)

# Approximate nadir point: the worst (maximum) value for each objective in the final generation
approx_nadir_all_population = res.history[len(res.history)-1].pop.get("F").max(axis=0)

# Normalize the objective values to a [0,1] scale
nF_all_population = (res.history[len(res.history)-1].pop.get("F") - approx_ideal_all_population) / (approx_nadir_all_population - approx_ideal_all_population)

# Define the reference point slightly beyond the worst normalized values
ref_point = nF_all_population.max(axis=0) + 0.1

# Compute the Hypervolume (HV) indicator
ind = HV(ref_point=ref_point)
print("HV =", np.round(ind(nF),2))  # Print the calculated hypervolume value
```

In my case, the ***hypervolume*** value is ***0.69***, which is a ***good result***, considering that our ***study area contains "empty" regions*** where no solutions from the ***Pareto front*** are present. This is despite the fact that the ***reference directions were evenly distributed across the entire search space***. This indicates that these ***empty regions are not suitable*** for offshore wind farm placement, as they are ***dominated (outperformed) by other, more optimal solutions***.

### **Scaling Issues in Multi-Criteria Optimization**

In multi-objective optimization, when one of the objective values has a significantly smaller range (spread of values) compared to others, prioritizing this criterion using weights can lead to undesirable results. This effect is observed in the problem of selecting the optimal solution from the ***Pareto front***, where ***wind speed has a much smaller range*** than ***water depth and distance to port***. As a result, when increasing the weight of wind speed, the other parameters ***deteriorate disproportionately***, since the optimizer can easily modify the larger-scale values (depth and distance) but has limited influence on wind speed.


When using the ***pseudo-weights method (`PseudoWeights`)***, the spread of objective functions is ***not accounted for by default***. As a result, if the weight of wind speed increases, the optimization algorithm attempts to improve this metric. However, since its ***spread is too small***, improvements are minimal, while depth and distance ***vary significantly***, leading to their excessive deterioration.
